```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC


iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]
y = (iris["target"] == 2).astype(np.float64)

print(y)

svm_cls = Pipeline((("scaler", StandardScaler()), 
                   ("linear_svr", LinearSVC(C=1, loss="hinge"))))
svm_cls.fit(X, y)

svm_cls.predict([[5.5, 1.7]])
```

```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt


%matplotlib inline

plt.figure(figsize=(16, 9))
X, y = make_moons(n_samples=1000,noise=0.1)
plt.title('make_moons function example')
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)

polynomial_svm_cls = Pipeline((("poly_features", PolynomialFeatures(degree=3)), 
                               ("scaler", StandardScaler()), 
                               ("svm_cls", LinearSVC(C=10, loss="hinge"))))
polynomial_svm_cls.fit(X, y)

polynomial_svm_cls.predict([[1.0, -0.5]])
polynomial_svm_cls.predict([[1.0, 0.5]])
```

```python
from sklearn.svm import SVC


poly_kernel_svm_clf = Pipeline((("scaler", StandardScaler()), 
                                ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))))
poly_kernel_svm_clf.fit(X, y)
polynomial_svm_cls.predict([[1.0, 0.5]])
```

```python
rbf_kernel_svm_clf = Pipeline((("scaler", StandardScaler()), 
                                ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))))
rbf_kernel_svm_clf.fit(X, y)
rbf_kernel_svm_clf.predict([[1.0, 0.5]])
```

```python
from sklearn.svm import LinearSVR

svm_reg = LinearSVR(epsilon=1.5)
svm_reg.fit(X, y)
```

```python
from sklearn.svm import SVR

svm_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.1)
svm_reg.fit(X, y)
```
